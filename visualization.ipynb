```python
import matplotlib.pyplot as plt
import lesion
import torch
import torchvision
import torch.nn as nn
from torch.utils.data import Dataset, DataLoader
from torchvision import transforms, utils
import numpy as np
import unet
import residual_unet
import utils as util
import torch.nn.functional as F
from sklearn.metrics import jaccard_score as jsc
device = torch.device("cuda:0")
import ECE
```

```python
class ToTensor_segmap(object):
    """Convert ndarrays in sample to Tensors."""

    def __call__(self, segmap):
        # swap color axis because
        # numpy image: H x W x C
        # torch image: C X H X W
        # In this case, Channels is 1, so there is no need to swap since data is in HxW      
  
        segmap = np.array(segmap)
        return torch.from_numpy(segmap) / 255

image_transform = transforms.Compose([
        transforms.RandomGrayscale(1),
        transforms.ToTensor(),
        transforms.Normalize(mean=[0.485, 0.456, 0.406],
                                     std=[0.229, 0.224, 0.225])
    ])
seg_transform = ToTensor_segmap()
```

```python
batch_size = 2
dataset = lesion.LesionDataset("data",folder_name = 'val',joint_transform=False,img_transform=image_transform, seg_transform=seg_transform,verbose = True)
loader = DataLoader(dataset, batch_size=batch_size,shuffle=False, num_workers=4)
```

```python

model1 = torch.load("models/" + "gate_all.pth")
model2 =torch.load("models/" + "gate_all_1.pth")
model3 =torch.load("models/" + "gate_all_2.pth")
model4 =torch.load("models/" + "gate_all_3.pth")
model5 =torch.load("models/" + "gate_all_4.pth")

model1.eval()
model2.eval()
model3.eval()
model4.eval()
model5.eval()
```

```python
def apply_dropout(m):
    if type(m) == nn.Dropout2d:
        m.train()
        
#T is the number of times to sample to approximate the posterior predictive distribution
#Assume model in eval mode
def pred_mean(loader,model,T,bayesian = True):
    total = len(loader)*batch_size
    if(bayesian):
        print("Bayesian Mode")
        model.apply(apply_dropout)
    preds = np.zeros((total, 2, 512, 1024))
    var = np.zeros((total, 2, 512, 1024))
    MI = np.zeros((total,512, 1024))
    targets = np.zeros((total, 512, 1024))
    for iteration in range(T):
        for i,data in enumerate(loader):
                image = data[0].to(device)
                m = nn.Softmax2d()
                output = model(image)
                output = m(output)
                output =  output.detach().cpu().numpy()
                segmap = data[1].numpy()
                preds[batch_size*i:batch_size*i + batch_size] += output
                MI[batch_size*i:batch_size*i + batch_size] += output[:,0,]*np.log(output[:,0,])+(output[:,1,]*np.log(output[:,1,]))
                if(iteration==0):
                    targets[batch_size*i:batch_size*i + batch_size] = segmap
    mean = preds / T
    
    entropy = -1*(mean[:,0,]*np.log(mean[:,0,]) + ((mean[:,1,])*np.log(mean[:,1,])))
    MI = entropy + (MI / T)
    return mean,entropy,MI,targets
    
    
    

    
     
    
```

```python
pred_temp = None
entropy_temp = None
mi_temp = None

pred,entropy,MI,targets = pred_mean(loader,model1,5)
pred_temp = pred
entropy_temp = entropy
mi_temp = MI

pred,entropy,MI,_= pred_mean(loader,model2,5)
pred_temp += pred
entropy_temp += entropy
mi_temp += MI

pred,entropy,MI,_=  pred_mean(loader,model3,5)
pred_temp += pred
entropy_temp += entropy
mi_temp += MI


pred,entropy,MI,_=  pred_mean(loader,model4,5)
pred_temp += pred
entropy_temp += entropy
mi_temp += MI


pred,entropy,MI,_= pred_mean(loader,model5,5)
pred_temp += pred
entropy_temp += entropy
mi_temp += MI
```

```python
pred_temp/=5
entropy_temp /=5
mi_temp /= 5
```

```python
confidence = pred_temp.max(axis=1)
avg_conf = 0
for i in range(150):
    avg_conf += confidence[i].mean()
avg_conf/=150
avg_conf
```

```python
avg_uncertainty = 0 #refers to predictive uncertainty
avg_mi = 0
#150 images in val set
for i in range(150):
    avg_uncertainty+= entropy_temp[i].mean()
    avg_mi += mi_temp[i].mean()
avg_uncertainty /=150
avg_mi /=150
print("Avg Predictive Uncertainty:{} Avg Mutual Information:{} ".format(avg_uncertainty,avg_mi))
```

```python
batch_size = 2
dataset = lesion.LesionDataset("data",folder_name = 'test',joint_transform=False,img_transform=image_transform, seg_transform=seg_transform,verbose = True)
loader = DataLoader(dataset, batch_size=batch_size,shuffle=False, num_workers=4)
```

```python
pred_temp = None
entropy_temp = None
mi_temp = None

pred,entropy,MI,targets = pred_mean(loader,model1,5)
pred_temp = pred
entropy_temp = entropy
mi_temp = MI

pred,entropy,MI,_= pred_mean(loader,model2,5)
pred_temp += pred
entropy_temp += entropy
mi_temp += MI

pred,entropy,MI,_=  pred_mean(loader,model3,5)
pred_temp += pred
entropy_temp += entropy
mi_temp += MI


pred,entropy,MI,_=  pred_mean(loader,model4,5)
pred_temp += pred
entropy_temp += entropy
mi_temp += MI


pred,entropy,MI,_= pred_mean(loader,model5,5)
pred_temp += pred
entropy_temp += entropy
mi_temp += MI

pred_temp/=5
entropy_temp /=5
mi_temp /= 5
```

```python
pred = pred_temp.argmax(axis=1) #test prediction
cm_predictive = np.copy(entropy_temp)  #certainty map using predictive uncertainty
cm_mi = np.copy(mi_temp)  #certainty map using mutual information
upper_predictive = np.nonzero(cm_predictive > avg_uncertainty) 
lower_predictive = np.nonzero(cm_predictive <= avg_uncertainty)
upper_mi=np.nonzero(cm_mi > avg_mi)
lower_mi = np.nonzero(cm_mi <= avg_mi)

cm_predictive[upper_predictive] = 0
cm_predictive[lower_predictive]=1
cm_mi[upper_mi] = 0
cm_mi[lower_mi] = 1
```

```python

def get_score(pred,t):
    jaccard = 0
    total = len(loader)*batch_size
    for i in range(total):
        jaccard += jsc(pred[i].reshape(-1),t[i].reshape(-1))
    jaccard /= total
    
    return jaccard,np.count_nonzero(pred==t)/pred.reshape(-1).shape[0]
        
```

```python
jaccard,accuracy =get_score(pred,targets)
print("Jaccard Score:{},  Accuracy:{} ".format(jaccard,accuracy))
```

```python
'''
P(accurate|certain)
'''
def ac(gt,pred,cm):
    #gt = groundtruth, pred = predictions, cm = certainty map
    loc = np.nonzero(cm==1) # location of certain pixels
    gt = gt[loc]
    pred = pred[loc]
    p = np.count_nonzero(gt == pred) # number of accurate pixels given they are certain
    return p/np.count_nonzero(cm==1)
'''
p(uncertain|inaccurate)
'''
def uia(gt,pred,cm):
    loc = np.nonzero(gt != pred) # location of inaccurate pixels
    cm = cm[loc]
    return np.count_nonzero(cm==0)/np.count_nonzero(gt!=pred)
```

```python
pac = ac(targets,pred,cm_predictive) #prob of accurate given certain
puia =uia(targets,pred,cm_predictive) #prob of uncertain given inaccurate
```

```python
print("Prob of accurate given certain is:{} ".format(pac))
print("Prob of uncertain given inaccurate is:{} ".format(puia))
```
